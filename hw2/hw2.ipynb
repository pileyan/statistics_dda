# Задание 1: Бутстреп "не по юниту рандомизации"

В папке лежит файл ***orders_2x20000_users_unique_ids.csv***. Там лежат данные про заказы и пользователей. Сравните показатель "средний чек" в группах с помощью бутстрепа двумя способами:

    1) Агрегируйте заказы до уровня пользователя и семплируйте пользователей, а потом уже считайте средние чеки по группам. 
    2) Напрямую семплируйте заказы.

Сравните полученные доверительные интервалы на разницу. Какой из вариантов правильный и почему?

# Задание 3: проверка границ применимости тестов

Используя метод, который был показан на последней лекции, проверьте, насколько тесты устойчивы к "перемешиванию" в группах. Допустим, что вы измеряете CTR и находитесь в области с "нормальной" работоспособностью тестов (пользователи не слишком различные и просмотры не слишком скошены), но часть наблюдений в группах "пересекается" – одни и те же пользователи с одними и теми же измерениями в обеих группах. Оцените, при каком проценте пересечения между группами корректность тестов нарушается? Предлагается проверить:

    1) t-тест на кликах
    2) Бакетное преобразование aka Bag of Little Bootstraps
    3) Линеаризацию
    4) Дельта-метод 

А есть ли между ними разница с точки зрения устойчивости к "пересечению" групп?
